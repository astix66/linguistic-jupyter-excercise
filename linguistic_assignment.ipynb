# CODING AND DESCRIPTIONS

```python
from bs4 import BeautifulSoup as bs
import requests
import string

link = requests.get('https://www.foxnews.com/lifestyle/exposing-trans-agenda-aimed-kids-faith-leader-reveals-how-parents-can-keep-children-safe')
parsed = bs(link.content, 'html.parser')

text = parsed.find_all('p')

print(text)
```

▶ This piece of ***exceptional*** American journalism has been selected due to Juptyer's issue with certificate valdation, which inexplicably allows to easily access Fox News (out of several tested media outlets)...

▶ As for the preceiding code, it extracts the text and strips it from white spaces, which in turn ensures better clarity.

```python
stringed = []

for s in text:
  if len(s.getText().strip()) > 0:
    stringed.append(s.getText())

print(stringed)
```

▶ Returns the previously accessed text, but "stringed." 🎆

```python
translator = str.maketrans('', '', string.punctuation)
stringed_no_punctuation = [s.translate(translator) for s in stringed]
print(stringed_no_punctuation)
```

▶ This code creates a translation table that manages to remove *most* of the punctuation marks; however, a number of odd ones out remain.

```python
import nltk
nltk.download("all")
```

```python
joined = ''.join(map(str,stringed_no_punctuation))
tokenized = nltk.word_tokenize(joined)
print(tokenized)
```

▶ This code employs the nltk module to turn the text into tokens.

▶ Parts of speech tagging can be employed on the tokens, like this:

```python
pos_tagged = nltk.pos_tag(tokenized)
print(*pos_tagged, sep='\n')
```

▶ As you can see, the tokenized variable (as well as its pos_tagged counterpart) contains loose apostrophes or dashes. These will be ignored, but not forgotten by perfectionists.

▶ The pos_tagged variable has been separated by newline for clarity!

![Very furious kitten](https://i.pinimg.com/564x/e9/13/f1/e913f1e8da6926a456e0d5e271b9878c.jpg)

```python
tags = [tag for word, tag in pos_tagged]

freq_dist = nltk.FreqDist(tags)
print(freq_dist.most_common())
```

▶ This code analyzes the frequency distribution of the POS tags. Let's try to visualize it using different modules!

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df = pd.DataFrame(list(freq_dist.items()), columns=['POS Tag', 'Count'])
df_sorted = df.sort_values(by='Count', ascending=False)
plt.figure(figsize=(16, 6))
palette = sns.color_palette('husl', len(df_sorted))
sns.barplot(x='POS Tag', y='Count', data=df_sorted, hue='POS Tag', palette=palette, dodge=False, legend=False)
plt.xticks(rotation=45)
plt.title('POS Tag Frequency Distribution')
plt.show()
```

▶ Let us utilize nltk one last time to check for word frequency!

```python
word_frequency = nltk.FreqDist(tokenized)

sorted_frequency = dict(sorted(word_frequency.items(), key=lambda x:x[1], reverse=True))

print('Word Frequency Distribution:')
for w, f in sorted_frequency.items():
    print(f'{w}: {f}')
```

▶ This operation displays a notable feature of the English language: most commonly used words are prepositions and a determiner.

▶ Additionally, the search proves that the tokenized text includes an extensive number of typos (lack of space between words).
